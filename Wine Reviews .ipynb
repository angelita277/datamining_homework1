```python
datafile_path='/kaggle/input/wine-reviews/winemag-data_first150k.csv'
datafile=pd.read_csv(datafile_path)
datafile.head(10)
```

# 3.1数据可视化和摘要
## 数据摘要
### 标称属性，给出每个可能取值的频数
第一列为标号，不考虑

```python

columns=['country','description','designation','points','price','province','region_1','region_2','variety','winery']
for column in columns:
    print(datafile[column].value_counts())
    print('-'*40)
```

### 数值属性，给出5数概括及缺失值的个数

```python
print(datafile.describe())
print('-'*80)
print(datafile.isnull().sum())
```

5数概括即为第一表格中最后五行，即最小值(min)，第一四分位数(25%)，中位数(50%)，第三四分位数(75%)，最大值(max)。
第二个表为确实数据

## 数据可视化
### 使用直方图、盒图等检查数据分布及离群点
由于只有'points','price'是数值，因此仅对这2列进行处理。
如下所示为直方图

```python
import matplotlib.pyplot as plt
for x in ['points','price']:
    dt1=pd.DataFrame(datafile[x],columns=[x])
    dt1.plot.hist(title=x,bins=100)
    dt2=pd.DataFrame(datafile[x],columns=[x])
    dt2.plot.box()
```

## 3.2数据缺失的处理
### 将缺失部分剔除

```python
data_dropna=datafile.dropna(axis=0,how='any')
print('原数据有'+str(datafile.shape[0])+'行')
print('新数据有'+str(data_dropna.shape[0])+'行')
print("删除了"+str(datafile.shape[0]-data_dropna.shape[0])+"行")
```

### 用最高频率值来填补缺失值
填补为none

```python
datafile.isnull()=datafile.isnull().fillna('None')
print(datafile.isnull().sum())
```
